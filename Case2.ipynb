```python
#Вам нужно написать программу, которая принимает на вход файл с данными об игроках и их результатах в одной из игр,
#выводит на экран рейтинг игроков по убыванию их среднего балла.
#Кроме того, программа должна строить круговую диаграмму игроков по среднему баллу.
```

```python
#Подключение библиотек

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as mb
```

```python
#Чтение файла
game = pd.read_csv('players.csv', delimiter=',')
print(game)
```

```python
#Нахождение среднего балла(score/games)
aa = float(game.loc[game['name'] == 'Alice', 'score'] / game.loc[game['name'] == 'Alice', 'games'])
bb = float(game.loc[game['name'] == 'Bob', 'score'] / game.loc[game['name'] == 'Bob', 'games'])
cc = float(game.loc[game['name'] == 'Charlie', 'score'] / game.loc[game['name'] == 'Charlie', 'games'])
dd = float(game.loc[game['name'] == 'David', 'score'] / game.loc[game['name'] == 'David', 'games'])
ee = float(game.loc[game['name'] == 'Eve', 'score'] / game.loc[game['name'] == 'Eve', 'games'])

#Создание дополнительной колонки
game["mean_score"] = [aa, bb, cc, dd , ee]
game

#Сортировка по среднему баллу
game.sort_values(by='mean_score', ascending=True)
```

```python
#Создание графика

#Создание списков
gamers = ["Alice", "Bob", "Charlie", "David", "Eve"]
scores = [aa,bb,cc,dd,ee]

# создание круговой диаграммы
mb.pie(scores, labels=gamers, autopct="%1.1f%%")
mb.title("Процент среднего количества баллов у игроков")
mb.show()
```
